# Importing the required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import os
from glob import glob
from tqdm import tqdm, tqdm_notebook
import random
from PIL import Image

import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.applications import *
from tensorflow.keras.callbacks import *
from tensorflow.keras.initializers import *
from tensorflow.keras.preprocessing.image import ImageDataGenerator

from numpy.random import seed
```

# Importing the data

```python
list_files = os.listdir('../input/skin-cancer-mnist-ham10000/')
for file in list_files:
    print(file)
```

## Creating a dictionary to store all the image path w.r.t to image_id

```python
path = '../input/skin-cancer-mnist-ham10000/'
images_path = {os.path.splitext(os.path.basename(x))[0]: x for x in glob(os.path.join(path, '*', '*.jpg'))}

count = 0
for k,v in images_path.items():
    print(k, v)
    count += 1
    if count > 5:
        break
```

## Creating a Dictinary to store the full-forms of entries in dx

```python
lesion_type_dict = {
    'nv': 'Melanocytic nevi',
    'mel': 'Melanoma',
    'bkl': 'Benign keratosis-like lesions ',
    'bcc': 'Basal cell carcinoma',
    'akiec': 'Actinic keratoses',
    'vasc': 'Vascular lesions',
    'df': 'Dermatofibroma'
}
```

```python
skin_df = pd.read_csv(os.path.join(path,'HAM10000_metadata.csv'))
skin_df.head()
```

## Creating :
* A column to store img path
* A column to store the full forms of dx
* A column to store the encodings of cell_type or dx

```python
skin_df['path'] = skin_df['image_id'].map(images_path.get)
skin_df['cell_type'] = skin_df['dx'].map(lesion_type_dict.get)
skin_df['cell_type_idx'] = pd.Categorical(skin_df['cell_type']).codes
skin_df.head()
```

```python
skin_df.info()
```

```python
skin_df.describe()
```

## Creating another column to store the images as arrays of shape (71,71,3) 
                                        

```python
skin_df['image'] = skin_df['path'].map(lambda x: np.asarray(Image.open(x).resize((71,71))))
skin_df.head()
```

## Checking the image size distribution

```python
skin_df['image'].map(lambda x: x.shape).value_counts()
```

## X = Images; y = Encoding of type of Cancer

```python
X = skin_df['image']
y = skin_df['cell_type_idx']
```

## Data Normalization

```python
X = X.values
X = X/255
X.shape
```

## Reshaping X so as to set the no. of examples or batch as the first entry

```python
lst = []
for _ in X:
    lst.append(_)
X = np.array(lst)
print(X.shape)
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.20,random_state=28)
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train, test_size=0.20,random_state=28)

print(f"Shape of x_train : {x_train.shape}, Shape of y_train : {y_train.shape}")
print(f"Shape of x_train : {x_val.shape}, Shape of y_train : {y_val.shape}")
print(f"Shape of x_train : {x_test.shape}, Shape of y_train : {y_test.shape}")
```

## One hot encoding of the y or labels

```python
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train, num_classes = 7)
y_val = to_categorical(y_val , num_classes=7)

print(f"Shape of y_train : {y_train.shape}")
print(f"Shape of y_val : {y_val.shape}")
```

## Training Parameters

```python
batch_size = 256
train_input_shape = (71, 71, 3)
n_classes = 7
```

# Building a model

* Using Xception NN as our base model 
* Then we will add some custom layers to our base model
* We will use EarlyStopping and ReduceLROnPlateau as our callbacks to monitor validation loss

* But this will not be our final model. We will use this model's 7th layer's output as features for our XGBOOST model.

```python
base_model = Xception(include_top = False , input_shape = train_input_shape, weights=None)

for layer in base_model.layers:
    layer.trainable = True
```

```python
# Adding layers to the base model

model = base_model.output
model = Flatten()(model)

model = Dense(512, kernel_initializer='he_uniform')(model)
model = Dropout(0.2)(model)
model = BatchNormalization()(model)
model = Activation('relu')(model)

model = Dense(128, kernel_initializer='he_uniform')(model)
model = Dropout(0.2)(model)
model = BatchNormalization()(model)
model = Activation('relu')(model)

model = Dense(52, kernel_initializer='he_uniform')(model)
model = Dropout(0.2)(model)
model = BatchNormalization()(model)
model = Activation('relu')(model)

model = Dense(16, kernel_initializer='he_uniform')(model)
model = Dropout(0.2)(model)
model = BatchNormalization()(model)
model = Activation('relu')(model)

output = Dense(n_classes, activation='softmax')(model)

model = Model(inputs=base_model.input, outputs=output)
```

```python
optimizer = Adam(lr=0.0001)
model.compile(loss='categorical_crossentropy',
              optimizer=optimizer, 
              metrics=['accuracy'])

model.summary()
```

```python
n_epoch = 10

early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=1, 
                           mode='auto', restore_best_weights=True)

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=5, 
                              verbose=1, mode='auto')
```

# Training the Model

```python
history = model.fit(x_train,
                    y_train,
                    epochs=60,
                    callbacks=[reduce_lr,early_stop],
                    validation_data=(x_val,y_val))
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,7))

ax[0].plot(history.history['loss'], color='b', label="Training loss")
ax[0].plot(history.history['val_loss'], color='r', label="validation loss",axes =ax[0])
legend = ax[0].legend(loc='best', shadow=True); ax[0].grid()

ax[1].plot(history.history['accuracy'], color='b', label="Training accuracy")
ax[1].plot(history.history['val_accuracy'], color='r',label="Validation accuracy")
legend = ax[1].legend(loc='best', shadow=True); ax[1].grid()
```
